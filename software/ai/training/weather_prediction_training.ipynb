```python
def create_supervised_dataset(df, window_size=6):
    X = []
    y = []
    target_datetimes = []

    for i in range(window_size - 1, len(df) - 1):
        row_features = {}

        # Historique sur la fenêtre glissante
        for j in range(window_size):
            idx = i - window_size + 1 + j
            lag = window_size - 1 - j  # pour nommer t-5, t-4, ..., t

            if lag > 0:
                suffix = f"t-{lag}"
            else:
                suffix = "t"

            row_features[f"TEMP_{suffix}"] = df.iloc[idx]["TEMP"]
            row_features[f"HUMI_{suffix}"] = df.iloc[idx]["HUMI"]
            row_features[f"PRES_{suffix}"] = df.iloc[idx]["PRES"]

        # Features temporelles de l'instant à prédire : t+1
        next_idx = i + 1
        row_features["hour_sin_t+1"] = df.iloc[next_idx]["hour_sin"]
        row_features["hour_cos_t+1"] = df.iloc[next_idx]["hour_cos"]
        row_features["month_sin_t+1"] = df.iloc[next_idx]["month_sin"]
        row_features["month_cos_t+1"] = df.iloc[next_idx]["month_cos"]

        # Cible
        X.append(row_features)
        y.append(df.iloc[next_idx]["TEMP"])
        target_datetimes.append(df.iloc[next_idx]["datetime"])

    X_df = pd.DataFrame(X)
    y_series = pd.Series(y, name="TARGET_TEMP_t+1")
    dt_series = pd.Series(target_datetimes, name="target_datetime")

    return X_df, y_series, dt_series
```

```python
window_size = 6

X, y, target_dt = create_supervised_dataset(df, window_size=window_size)

print("Shape de X :", X.shape)
print("Shape de y :", y.shape)

display(X.head())
display(y.head())
display(target_dt.head())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_X = StandardScaler()
```

```python
X_train_scaled = scaler_X.fit_transform(X_train)
```

```python
X_val_scaled = scaler_X.transform(X_val)
```

```python
X_test_scaled = scaler_X.transform(X_test)
```

```python
np.mean(X_train_scaled)
```

```python
np.std(X_train_scaled)
```

```python
feature_names = X_train.columns.tolist()
scaler_means = scaler_X.mean_
scaler_stds = scaler_X.scale_
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
input_dim = X_train_scaled.shape[1]
```

```python
model = keras.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(1)
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True)
```

```python
history = model.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_val_scaled, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)
```

```python
len(history.history['loss'])
```

```python
test_loss, test_mae = model.evaluate(X_test_scaled, y_test, verbose=0)

print(f"Test Loss (MSE) : {test_loss:.4f}")
print(f"Test MAE        : {test_mae:.4f} °C")
```
